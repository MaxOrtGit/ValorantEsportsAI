```python
#sqllite
import sqlite3
import pandas as pd

files = ['savedataOS.db', 'savedataWP.db']

# from the matches table get all the matches, Get the "t1", "t2", "t1oo", "t2oo", and "date_winner"
def get_matches(files):
  df = pd.DataFrame()
  for file in files:
    conn = sqlite3.connect(file)
    query = "SELECT t1, t2, t1oo, t2oo, date_winner FROM match"
    df = df.append(pd.read_sql_query(query, conn))
    conn.close()
  return df
  
df = get_matches(files)
display(df)
```
